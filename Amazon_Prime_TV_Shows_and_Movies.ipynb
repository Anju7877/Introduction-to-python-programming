**DATA WRANGLING**

```python

# Import Libraries
import pandas as pd
import numpy as np
import ast

# 1. Load Dataset

Amazon_titles = pd.read_csv("titles.csv")
Amazon_credits = pd.read_csv("credits.csv")

# 2. Check Shape

print("Titles Shape :", Amazon_titles.shape)
print("Credits Shape:", Amazon_credits.shape)

# 3. Remove Duplicate Rows

Amazon_titles.drop_duplicates(inplace=True)
Amazon_credits.drop_duplicates(inplace=True)


# 4. Drop Unnecessary Columns

# character column has many null values and low business use
Amazon_credits.drop('character', axis=1, inplace=True)


# 5. Handle Missing Values

# Seasons null means Movie
Amazon_titles['seasons'] = Amazon_titles['seasons'].fillna(0)

# Numeric Columns
Amazon_titles['imdb_score'] = Amazon_titles['imdb_score'].fillna(0)
Amazon_titles['imdb_votes'] = Amazon_titles['imdb_votes'].fillna(0)
Amazon_titles['tmdb_score'] = Amazon_titles['tmdb_score'].fillna(0)
Amazon_titles['tmdb_popularity'] = Amazon_titles['tmdb_popularity'].fillna(0)

# Text Columns
Amazon_titles['age_certification'] = Amazon_titles['age_certification'].fillna('Not Rated')
Amazon_titles['description'] = Amazon_titles['description'].fillna('No Description')

# 6. Convert String Lists to Real Lists
def clean_list_column(row):

    if isinstance(row, str) == False:
        return ['Unknown']

    try:
        converted = ast.literal_eval(row)

        if len(converted) == 0:
            return ['Unknown']

        return converted

    except:
        return ['Unknown']

# Apply Function
Amazon_titles['genres'] = Amazon_titles['genres'].apply(clean_list_column)
Amazon_titles['production_countries'] = Amazon_titles['production_countries'].apply(clean_list_column)


# 7. Fix Data Types

Amazon_titles['seasons'] = Amazon_titles['seasons'].astype(int)
Amazon_titles['release_year'] = Amazon_titles['release_year'].astype(int)


# Decade Column
Amazon_titles['decade'] = (Amazon_titles['release_year'] // 10) * 10

# IMDb Popular Category
Amazon_titles['popular_content'] = np.where(
    Amazon_titles['imdb_votes'] > Amazon_titles['imdb_votes'].median(),
    'Popular',
    'Less Popular'
)

# 9. Final Info Check


print(Amazon_titles.info())
print(Amazon_credits.info())

# 10. Save Cleaned Files


Amazon_titles.to_csv("Amazon_titles_cleaned.csv", index=False)
Amazon_credits.to_csv("Amazon_credits_cleaned.csv", index=False)

print("Data Wrangling Completed Successfully!")
```

### What all manipulations have you done and insights you found?

1.Handling Duplicates: Removed duplicate rows from  both datasets(titles and credits) to ensure unique records.


2.Dropping Column: I dropped  the character column from the credits dataset because it had over 16,000 missing values, and teh specific role names were not required for our analysis of Actors and Directors.


3 Imputing Missing Values:
* seasons: Filled missing values with 0. This logically distinguishes Movies(0 seasons) from TV shows(1+ seasons).
* age_certification: Filled missing entries with " Not Rated" to preserve the data for analysis rather than deleting rows.   
* Ratings: Filled missing scores with 0 to handle unrated content.

4.Data Type Conversion:

* Converted genres and production_countries from strings to actual lists.  
* Handled Empty Lists: Replaced empty entries[] in production_countries with [unknown] to avoid errors during visualization and analysis.

CHART-9 Distribution of Runtime(Movies VS.TV SHOWS)

```python
# Chart-9: Runtime Distribution (Movies vs TV Shows) - Final Working Code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
titles = pd.read_csv("titles.csv")

# Check correct column name (MOVIE / SHOW type column)
# Usually it is 'type' in this dataset

# Drop missing values
titles = titles.dropna(subset=['runtime'])

# Plot
plt.figure(figsize=(12,6))

sns.histplot(
    data=titles,
    x='runtime',
    hue='type',   # correct column
    bins=30,
    kde=True,
    element="step"
)

plt.title("Runtime Distribution: Movies vs TV Shows")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Count")

plt.show()
```

CHART 11: Average IMDb Score Over Time (2000-2021)

```python
# Average IMDb Score Over Time (2000-2021) - Alternative Code

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
titles = pd.read_csv("titles.csv")

# Keep required data only
df = titles[['release_year', 'imdb_score']].dropna()

# Filter years
df = df[(df['release_year'] >= 2000) & (df['release_year'] <= 2021)]

# Calculate yearly average IMDb score
avg_imdb = df.groupby('release_year')['imdb_score'].mean()

# Plot
plt.figure(figsize=(12,6))
plt.plot(avg_imdb.index, avg_imdb.values, marker='o', linewidth=2)

plt.title("Average IMDb Score Over Time (2000-2021)")
plt.xlabel("Release Year")
plt.ylabel("Average IMDb Score")
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read file
data = pd.read_csv("titles.csv")

# Pick numerical columns
num_cols = data.select_dtypes(include=["int64", "float64"])

# Create correlation table
correlation = num_cols.corr()

# Figure size
plt.figure(figsize=(12,8))

# Draw heatmap
sns.heatmap(
    correlation,
    annot=True,
    cmap="YlGnBu",
    square=True,
    fmt=".2f",
    cbar=True
)

# Title
plt.title("Feature Correlation Heatmap", fontsize=16)

plt.tight_layout()
plt.show()
```
